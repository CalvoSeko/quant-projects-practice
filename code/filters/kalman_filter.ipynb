# Kalman filter #
reference to: https://web.mit.edu/kirtley/kirtley/binlustuff/literature/control/Kalman%20filter.pdf
## Intro ##
Optimal solution for many tracking and data prediction tasks.
## Example ##
For minimising MSE in:
$$ y_k = a_k * x_k + n_k $$
where a is a gain term, x the information and n is the noise.
Objective: estimate $$x_k$$
